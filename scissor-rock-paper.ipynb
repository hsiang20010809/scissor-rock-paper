```python
import random
scissor = '''
         _______
     ---'   ____)__
             ______)
         __________)
         (____)
     ---.__(___)
     '''
rock = '''
        _______
    ---'   ____)
        (_____)
        (_____)
        (____)
    ---.__(___)
    '''
paper = '''
        _______
    ---'   ____)__
            ______)
            _______)
            _______)
    ---.__________)
    '''

a = 0
while a < 100:
    word = input("input your choice (s, r, p):").lower()
    if word == "0":
        break
    print("YOU:")
    if word == "s":
        print("scissor")
        print(scissor)
    elif word == "r":
        print("rock")
        print(rock)
    else:
        print("paper")
        print(paper)
    pc = "scissor, rock, paper"
    pc = pc.split(", ")
    pc_choice = random.randint(0, 2)
    print("PC:")
    if pc_choice == 0:
        print("scissor")
        print(scissor)
    elif pc_choice == 1:
        print("rock")
        print(rock)
    else:
        print("paper")
        print(paper)
    if word == "s" :
        if pc_choice == 0:
            print("peace")
        elif pc_choice == 1:
            print("you lose")
        else:
            print("you win")
    if word == "r" :
        if pc_choice == 0:
            print("you win")
        elif pc_choice == 1:
            print("peace")
        else:
            print("you lose")
    if word == "p" :
        if pc_choice == 0:
            print("you lose")
        elif pc_choice == 1:
            print("you win")
        else:
            print("peace")
```
